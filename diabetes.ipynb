```python
from pandas.core.algorithms import value_counts
from numpy import number
def showPie(data,number): #함수
 data_true=(data>number).value_counts(normalize=True)[True]
 data_false=(data>number).value_counts(normalize=True)[False]
 data_set=[data_true,data_false]
 categories=["True","False"]
 plt.pie(data_set, labels=categories, autopct="%0.1f")
```

```python
df_zero_to_mean=df #0으로 된 값 찾기 
```

```python
import numpy as np
df_zero_to_mean["Glucose"].replace(0,np.NaN,inplace=True)
df_zero_to_mean["BloodPressure"].replace(0,np.NaN,inplace=True)
df_zero_to_mean["SkinThickness"].replace(0,np.NaN,inplace=True)
df_zero_to_mean["Insulin"].replace(0,np.NaN,inplace=True)
df_zero_to_mean["BMI"].replace(0,np.NaN,inplace=True)
df_zero_to_mean["DiabetesPedigreeFunction"].replace(0,np.NaN,inplace=True)
```

```python
df_zero_to_mean.isnull().sum()
```

```python
df_zero_to_mean["Glucose"].fillna(df_zero_to_mean["Glucose"].mean(),inplace=True)   #데이터 평균값으로 바꿈
df_zero_to_mean["BloodPressure"].fillna(df_zero_to_mean["Glucose"].mean(),inplace=True)
df_zero_to_mean["SkinThickness"].fillna(df_zero_to_mean["Glucose"].mean(),inplace=True)
df_zero_to_mean["Insulin"].fillna(df_zero_to_mean["Glucose"].mean(),inplace=True)
df_zero_to_mean["BMI"].fillna(df_zero_to_mean["Glucose"].mean(),inplace=True)
```
